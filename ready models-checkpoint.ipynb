```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
```

### DB SOURCE - https://www.kaggle.com/anthonypino/melbourne-housing-market/

```python
df = pd.read_csv(r"C:\Users\user\Desktop\machine learning\house_pricing\Melbourne_housing_FULL.csv")
```

## previously removed some columns wich we dont need here, you can check full table ot the link(we dont need to overweight the model with excess data)

```python
df.tail(7)

```

# cleaning the db from NaNs

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
df.head(10) #check
```

```python
df = pd.get_dummies(df,columns = ['Suburb','CouncilArea','Type'])
```

```python
df.head(5)
```

# taking independant and dependant columns for X & Y vectors

```python
X = df.drop('Price', axis = 1)
```

## price will be prediction 

```python
y = df['Price']
```

```python
y
```

```python
X ##check
```

## SPLITTING THE DB (and shuffle - !!important) 70-30%

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, shuffle=True)
```

```python
X_train ##check
```

### selecting alorithm - gradient boosting regressor(we need regression not classification in this model)

```python
model = ensemble.GradientBoostingRegressor(n_estimators=150,learning_rate=0.1,max_depth=30,min_samples_split=4,
                                          min_samples_leaf=6,max_features=0.6,loss='huber')
```

## train the model

```python
model.fit(X_train,y_train)
```

# evaluation 

```python
mae_train = mean_absolute_error(y_train, model.predict(X_train))
```

```python
print("%.2f"%mae_train)
```

```python
mae_test = mean_absolute_error(y_test, model.predict(X_test))
```

```python
print("%.2f"%mae_test)
```
